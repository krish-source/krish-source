```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import networkx as nx
```

```python
var1 = yf.Ticker('AMZN')
```

```python
amazon = var1.institutional_holders
```

```python
amazon
```

```python
amazon['comp'] = var1.ticker
```

```python
amazon
```

```python
var2 = yf.Ticker('GOOGL')
```

```python
google = var2.institutional_holders
```

```python
google
```

```python
google['comp'] = var2.ticker
```

```python
google
```

```python
combined = pd.concat([amazon, google])
```

```python
combined
```

```python
G = nx.from_pandas_edgelist(combined, 'Holder', 'comp')
```

```python
nx.draw(G, with_labels=True)
```
