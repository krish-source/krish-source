```python
stocks = pd.read_csv("us_stocks.csv", header= [0,1], index_col= [0], parse_dates= [0]).Close
```

```python
stocks.head()
```

```python
norm = stocks.div(stocks.iloc[0,:]).mul(100)
```

```python
norm.tail()
```

```python
norm = norm.drop(['FB'], axis=1)
```

```python
norm
```

```python
cf.set_config_file(offline=True)
```

```python
norm.iplot()
```

```python
cf.colors.scales()
```

```python
norm.iplot(kind="line", fill=True, colorscale="blues")
```

```python
cf.getThemes()
```

```python
norm.iplot(kind="line", fill=True, colorscale="rdylgn", theme="solar", title="US stocks", xTitle="Time", yTitle="Stock Price")
```

```python
norm[["AXP", "MCD"]].iplot(kind= "spread", fill=True,  theme="solar", title="AXP vs MCD", xTitle="Time", yTitle="Stockprice")
```

```python
stocks_2 = pd.read_csv("us_stocks.csv", header= [0,1], index_col= [0], parse_dates= [0])
```

```python
fb = stocks_2.swaplevel(axis=1).FB
```

```python
fb.dropna()
```

```python
fb.loc["5-2018":"12-2018"].iplot(kind="candle")
```

```python
fb.loc["5-2018":"12-2018"].iplot(kind="ohlc")
```

```python
qf = cf.QuantFig(df=fb.loc["5-2018":"12-2018"])
```

```python
type(qf)
```

```python
qf.iplot(title= "FB", name = "FB")
```

```python
qf.add_sma(periods=20)
```

```python
qf.iplot(title= "FB", name = "FB")
```

```python
qf.add_bollinger_bands(periods=20, boll_std= 2)
```

```python
qf.iplot(title= "FB", name = "FB")
```
